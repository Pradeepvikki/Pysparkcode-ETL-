```python
from pyspark.sql.functions import col, lower, create_map, lit, when

# dictionary for word â†’ number
num_map = {
    "zero": 0,
    "one": 1,
    "two": 2,
    "three": 3,
    "four": 4,
    "five": 5,
    "six": 6,
    "seven": 7,
    "eight": 8,
    "nine": 9,
    "ten": 10
}

# convert dictionary to PySpark map
mapping_list = []
for k, v in num_map.items():
    mapping_list += [lit(k), lit(v)]

mapping_expr = create_map(*mapping_list)

# final conversion logic for Qty
df = df.withColumn(
    "Qty",
    when(
        col("Qty").rlike("^[0-9]+$"),      # already numeric (1, 02, 3)
        col("Qty").cast("int")
    ).when(
        lower(col("Qty")).isin(list(num_map.keys())),  # word numbers
        mapping_expr[lower(col("Qty"))]
    ).otherwise(None)   # NaN or invalid
)
df.show()
```

#### here change the rupees into dollors

```python
from pyspark.sql.functions import col, regexp_replace, when

# conversion rate
usd_rate = 83  # 1 USD = 83 INR

df = df.withColumn(
    "Order_Amount$",

    # Step 1: Remove symbols (â‚¹, $, commas)
    regexp_replace(col("Order_Amount$"), r"[â‚¹$,]", "")
        .alias("Order_Amount$")
)

# Step 2: Convert to USD
df = df.withColumn(
    "Order_Amount$",
    when(
        col("Order_Amount$").rlike("(?i).*inr.*"),   # values containing 'INR'
        (regexp_replace(col("Order_Amount$"), "(?i)inr", "").cast("double") / usd_rate)
    )
    .when(
        col("Order_Amount$").rlike("(?i)^usd"),      # 'USD 30'
        regexp_replace(col("Order_Amount$"), "(?i)usd", "").cast("double")
    )
    .otherwise(col("Order_Amount$").cast("double"))  # already in dollars
)
df.show()
```

```python
from pyspark.sql.functions import round, col

df = df.withColumn(
    "Order_Amount$",
    round(col("Order_Amount$"), 2)
)
df.show()
```

#### Date format change

```python
from pyspark.sql.functions import to_date, coalesce, col, date_format

# Step 1: Convert mixed formats into a valid date type
df = df.withColumn(
    "Order_Date",
    coalesce(
        to_date(col("Order_Date"), "yyyy/MM/dd"),
        to_date(col("Order_Date"), "dd-MM-yyyy"),
        to_date(col("Order_Date"), "yyyy.MM.dd"),
        to_date(col("Order_Date"), "dd/MM/yyyy"),
        to_date(col("Order_Date"), "MM-dd-yyyy"),
        to_date(col("Order_Date"), "MM.dd.yyyy"),
        to_date(col("Order_Date"), "MMMM dd yyyy"),
        to_date(col("Order_Date"), "MMMM dd'th' yyyy")
    )
)

# Step 2: Format final output into yyyy/MM/dd
df = df.withColumn("Order_Date", date_format(col("Order_Date"), "yyyy/MM/dd"))
df.show()
```

```python
from pyspark.sql.functions import lower, col, when

df = df.withColumn(
    "Delivery_Status",
    when(lower(col("Delivery_Status")).isin("delivered", "deliverd", "delivrd"), "Delivered")
    .when(lower(col("Delivery_Status")) == "pending", "Pending")
    .when(lower(col("Delivery_Status")).isin("on the way", "ontheway"), "On The Way")
    .when(lower(col("Delivery_Status")).isin("onhold", "on hold"), "On Hold")
    .when(lower(col("Delivery_Status")) == "cancelled", "Cancelled")
    .when(lower(col("Delivery_Status")) == "delivering", "Delivering")
    .otherwise(None)
)
df.show()
```

```python
from pyspark.sql.functions import col, lower, when

df = df.withColumn(
    "Payment_Mode",
    when(lower(col("Payment_Mode")).isin("credit", "creditcard", "credit card", "credit_card"), "Credit Card")
    .when(lower(col("Payment_Mode")).isin("debit", "debit card", "debitcard"), "Debit Card")
    .when(lower(col("Payment_Mode")).isin("paypal", "pay_pal"), "PayPal")
    .when(lower(col("Payment_Mode")).isin("upi"), "UPI")
    .when(lower(col("Payment_Mode")).isin("cash"), "Cash")
    .otherwise(None)
)
df.show()
```

```python
from pyspark.sql.functions import col, regexp_replace, when

# Step 1: Basic cleanup for repeated symbols
df = df.withColumn(
    "Email",
    regexp_replace(col("Email"), r"@+", "@")
)

df = df.withColumn(
    "Email",
    regexp_replace(col("Email"), r"\.+", ".")
)

df = df.withColumn(
    "Email",
    regexp_replace(col("Email"), r"#", "@")
)

# Step 2: Add domain if missing (ex: john@ -> john@unknown.com)
df = df.withColumn(
    "Email",
    when(col("Email").rlike(r"^[^@]+@$"), regexp_replace(col("Email"), "@$", "@unknown.com"))
    .otherwise(col("Email"))
)

# Step 3: Add .com if domain extension missing (ex: user@gmail -> user@gmail.com)
df = df.withColumn(
    "Email",
    when(~col("Email").rlike(r"\.[a-zA-Z]{2,}$"), col("Email") + ".com")
    .otherwise(col("Email"))
)

df.show()
```

```python
from pyspark.sql.functions import col, regexp_replace, lower, trim, when

# Lowercase + remove special chars
df = df.withColumn(
    "Ship_Address",
    trim(lower(regexp_replace(col("Ship_Address"), r"[^a-zA-Z ]", "")))
)

# Mapping valid Indian cities
df = (
    df
    .withColumn(
        "Ship_Address",
        when(col("Ship_Address").contains("mumbai"), "Mumbai")
        .when(col("Ship_Address").contains("delhi"), "Delhi")
        .when(col("Ship_Address").contains("new delhi"), "New Delhi")
        .when(col("Ship_Address").contains("noida"), "Noida")
        .when(col("Ship_Address").contains("pune"), "Pune")
        .when(col("Ship_Address").contains("chennai"), "Chennai")
        .when(col("Ship_Address").contains("gurgaon"), "Gurgaon")
        .when(col("Ship_Address").contains("bangalore"), "Bangalore")
        .when(col("Ship_Address").contains("hyderabad"), "Hyderabad")
        .when(col("Ship_Address").contains("coimbatore"), "Coimbatore")
        .otherwise("Unknown")
    )
)
df.show()
```
